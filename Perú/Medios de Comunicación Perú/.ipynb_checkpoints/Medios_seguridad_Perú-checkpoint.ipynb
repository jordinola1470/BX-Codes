```python
import pandas as pd
import numpy as np
import seaborn as sns
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
```

```python
base_sitios = pd.read_excel('sitios_peruanos.xlsx')
base_sitios.head(7)
```

```python
base_cuentas = pd.read_excel('autores_peruanos.xlsx')
base_cuentas.head(7)
```

```python
base_pegada = base_cuentas.append(base_sitios)
base_pegada.head(7)
```

```python
base_pegada['% Seguridad'] = round(base_pegada['Seguridad']/base_pegada['Medios']*100,2)
base_pegada.head(7)
```

```python
base_grafica = base_pegada[base_pegada['% Seguridad'] > 0.0]
ax = sns.barplot(x="Autores", y="% Seguridad", data=base_grafica)
for item in ax.get_xticklabels():
    item.set_rotation(90)
```

```python
base_pegada.to_excel('Base_medios_sitios_seguridad.xlsx')
```
