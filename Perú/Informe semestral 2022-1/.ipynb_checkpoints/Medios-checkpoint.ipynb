```python
import numpy as np
import pandas as pd
```

```python
base_totales = pd.read_excel('Total_medios_perú.xlsx')
base_integr = pd.read_excel('Integración_medios_perú.xlsx')
```

```python
base_final = pd.merge(right=base_totales, left=base_integr, how='right', left_on='Author', right_on='Author')
base_final.fillna(inplace=True, value= 0)
base_final['% Integración'] = round(base_final['Integración']/base_final['Total'],3)
base_final
```

```python
base_final.to_excel('base_final.xlsx')
```
