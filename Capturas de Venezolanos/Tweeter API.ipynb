```python
import re
data_twitter['Capturas'] = 0
data_twitter['Migrantes'] = 0
data_twitter['Capturas Migrantes'] = 0
patterns_captura = [r'\s+captu\w+\s+',r'\s+judic\w+\s+',r'\s+apresa\w+\s+',r'\s+arrest\w+\s+',r'\s+aprisio\w+\s+',r'\s+aprehe\w+\s+',r'\s+deten\w+\s+',r'\s+desman\w+\s+']
patterns_migrantes = [r'\s+venez\w+\s+',r'\s+extranj\w+\s+',r'\s+migran\w+\s+']
for index, row in data_twitter.iterrows():
    for pattern in patterns_captura:
        if re.search(pattern, str(row['text'])):
            data_twitter.loc[index, 'Capturas'] = 1
            for pattern_2 in patterns_migrantes:
                if re.search(pattern_2, str(row['text'])):
                    data_twitter.loc[index, 'Capturas Migrantes'] = 1
                    break
            break
    for pattern_3 in patterns_migrantes:
        if re.search(pattern_3, str(row['text'])):
            data_twitter.loc[index, 'Migrantes'] = 1
            break
data_twitter.head(5)
```

```python
from datetime import datetime
import dateutil.parser as parser
data_twitter['mes'] = 0
data_twitter['dia'] = 0
data_twitter['mes_a√±o'] = 0
data_twitter['a√±o'] = 0
parser.parse("2022-11-07T04:33:30.000Z").day
for index, row in data_twitter.iterrows():
    data_twitter.loc[index, 'mes'] = parser.parse(data_twitter.loc[index, 'created_at']).month
    data_twitter.loc[index, 'dia'] = parser.parse(data_twitter.loc[index, 'created_at']).day
    data_twitter.loc[index, 'a√±o'] = parser.parse(data_twitter.loc[index, 'created_at']).year
    data_twitter.loc[index, 'mes_a√±o'] = str(parser.parse(data_twitter.loc[index, 'created_at']).month) + "_" + str(parser.parse(data_twitter.loc[index, 'created_at']).year)
data_twitter
```
