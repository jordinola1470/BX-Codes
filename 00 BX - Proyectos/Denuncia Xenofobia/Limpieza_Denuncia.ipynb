# Limpieza de Base de Menciones de Xenofobia
Este notebook explicado sirve como guía para la limpieza de un archivo ```.xlsx``` extraído de Brandwatch que captura la conversación de migrantes venezolanos para un periodo determinado. En la primera versión del documento se usa el Piloto Barómetro 2021

````python
# Definimos la ubicación del archivo
path = input('Escriba la ruta del archivo - ')
#Importamos la base de datos que descargamos con los tweets de brandwatch 
xenofobia=pd.read_excel(path)
del path
#path='/Users/diegoalejandrobermudezsierra/Downloads/Barometro_Migrante_Colombia.xlsx'
xenofobia.head(11)
````

````python
base=fn.limpieza_base(xenofobia)
base.head()
print("Tamaño de la base: " + str(base.shape))
````

````python
del_cols=["Query_Name", "Url", "Page_Type", "Language", "Country_Code", "Continent", "Continent_Code", 
          "Country", "Added", "Assignment", "Avatar", "Category_Details", 
          "Checked", "Display_URLs", "Expanded_URLs", "Facebook_Author_ID", "Facebook_Comments",
          "Facebook_Likes", "Facebook_Role", "Facebook_Shares", "Facebook_Subtype",
          "Full_Text", "Full_Name", "Instagram_Comments", "Instagram_Followers", "Instagram_Following",
          "Instagram_Likes", "Instagram_Interactions_Count", "Instagram_Posts",
          "Last_Assignment_Date", "Latitude", "Location_Name", "Longitude", "Media_Filter",
          "Media_URLs", "Original_Url", "Priority", "Short_URLs", "Resource_Id", "Starred",
          "Status", "Subtype", "Thread_Created_Date", "Thread_Id",
          "Thread_URL", "Twitter_Author_ID", "Twitter_Channel_Role", "Twitter_Reply_to", "Twitter_Retweet_of", 
          "Updated", "Blog_Name", "Copyright", "Item_Review", "Page_Type_Name", "Linkedin_Comments", 
          "Linkedin_Engagement", "Linkedin_Impressions", "Linkedin_Likes", "Linkedin_Shares", "Linkedin_Sponsored", 
          "Linkedin_Video_Views", "Parent_Blog_Name", "Parent_Post_Id", "Pub_Type", "Publisher_Sub_Type", "Rating", 
          "Reddit_Score", "Region", "Region_Code", "Root_Blog_Name", "Root_Post_Id", "Weblog_Title", "Title"]
try:
    base = base.drop(del_cols, axis=1)
except: print("Alguna columna a borrar no se encuentra en la base")
# "Tags" se sacaron de la eliminación
for col in base.columns: 
    print(col)
````

````python
#Eliminamos todos los @Usuario (i.e. @ClaudiaLopez)
base['Clean_Text']=base['Snippet'].replace('@\\w+', '', regex=True)
#Eliminamos todos los hashtags
base['Clean_Text']=base['Clean_Text'].replace('#\\w+', '', regex=True)
#Eliminamos los URL: Todo lo que empieza por 
base['Clean_Text']=base['Clean_Text'].replace(r"http\S+", '', regex=True)
#Eliminamos todos los signos de puntuación
base['Clean_Text']=base['Clean_Text'].replace(r'[^\w\s]+', '', regex=True)
#Eliminamos los asentos de las letras
base['Clean_Text']=base['Clean_Text'].str.normalize('NFKD')\
        .str.encode('ascii', errors='ignore')\
        .str.decode('utf-8')
#Eliminamos los dobles espacios
base['Clean_Text']=base['Clean_Text'].replace('\s+', ' ', regex=True).str.strip()
#Eliminamos las filas que sean RT (agregar , na=False si hay NaNs)
base=base[~base.Snippet.str.startswith('RT')].reset_index(drop=True)
#Eliminamos las letras repetivas dentro de las palabras
base['Clean_Text']=base['Clean_Text'].replace(r'(.)\1+', r'\1\1', regex=True)

base.head()
````
