```python
import pandas as pd
import numpy as np
```

```python
def unir(bases):
    i = 1
    final = pd.DataFrame()
    while i <= bases:
        j = str(i)
        est = "porc_bots_" + j + ".xlsx"
        df = pd.read_excel(est)
        final = final.append(df)
        print(final.shape[0])
        i += 1
    #final = final.set_index(final["ID"]).drop(["ID"], axis=1).fillna(0)
    return final
```

```python
bots = unir(12)
```

```python
df = pd.read_csv('exported_audience_2021-06-22_at_15.13.41_UTC.csv') 
df
```

```python
base_merge = pd.merge(df, bots, how='left', left_on='twitter handle', right_on='Autores')
base_merge
```

```python
final = base_merge.drop(columns = ['Autores'])
final
```

```python
final.to_excel('bots_base_entera.xlsx')
```
