```python
def separar(df,bases):
    count = df.shape[0]
    obs_base = round(count/bases)
    i = 0
    d = {}
    while i <= bases:
        start = i * obs_base 
        stop = (i + 1) * obs_base
        if stop > count:
            stop = count
        if i != 0:
            start = start + 1
        d[i] = pd.DataFrame(df.loc[start:stop]) 
        i += 1
    return d
```
