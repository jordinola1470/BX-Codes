```python
def separar(df,bases):
    count = df.shape[0]
    obs_base = round(count/bases)
    i = 0
    d = {}
    while i <= bases:
        start = i * obs_base 
        stop = (i + 1) * obs_base
        if stop > count:
            stop = count
        if i != 0:
            start = start + 1
        d[i] = pd.DataFrame(df.loc[start:stop]) 
        i += 1
    return d
```

```python
d = separar(df,34)
d
```

```python
i = 1
for key, value in d.items():
    base = value
    j = str(i)
    print(j)
    base.to_excel("estudiante" + "_" + j + ".xlsx")
    i += 1
```

```python
def unir(bases):
    i = 1
    final = pd.DataFrame()
    while i <= bases:
        j = str(i)
        est = "estudiante_" + j + ".xlsx"
        df = pd.read_excel(est)
        final = final.append(df)
        print(final.shape[0])
        i += 1
    final = final.set_index(final["ID"]).drop(["ID"], axis=1).fillna(0)
    return final
```

```python
final = unir(34)
```

```python
final
```

```python
final_copy = final
print("Los primeros tipos son: " + str(final_copy.dtypes))
print("----------Miremos que hay en las otras columnas raras-----------")
print("Para Unnamed:10 " + str(final_copy[final_copy['Unnamed: 10'] != 0.0]))
print("Para Unnamed:9 " + str(final_copy[final_copy['Unnamed: 9'] != 0.0]))
print("Para color: " + str(final_copy[final_copy['Color'] != 0.0]))
final_copy = final_copy.drop(["Unnamed: 10", "Unnamed: 9", "Color"], axis=1)
print("Para Unnamed:8 " + str(final_copy[final_copy['Unnamed: 8'] != 0.0]))
print("Para Unnamed:7 " + str(final_copy[final_copy['Unnamed: 7'] != 0.0]))
for col in final_copy.columns: 
    if col != 'Snippet':
        print("Estamos limpiando la columna: " + col)
        print(final_copy[final_copy[col].apply(lambda x: isinstance(x, str))])
        if col == 'Victima':
            # Limpiamos caracteres no n√∫mericos, solo se encontr√≥ "666m", se pasa su informaci√≥n a la columna tags
            final_copy['Tag 666 777'] = np.where(final_copy[col] == "666m", 1, 0)
            # Usamos map o ternary para limpiar dobles errores no n√∫mericos que se tratan igual
            final_copy[col] = final_copy[col].replace(regex={'\w+': 0, '\s+': 0, '|': 0})
            print(final_copy[col])
        if col == 'Victimario' or col == 'Xenofobia' or col == 'Denuncia' or col == 'Positiva' or col == 'Positivo':
            final_copy[col] = final_copy[col].map({" ":0, 0.0:0, 
                                                   0:0, 1.0:1, "*":0})
        if col == 'Unnamed: 7':
            # Limpiamos caracteres no n√∫mericos, solo se encontr√≥ *
            # Limpiamos y reordenamos extra√±os inputs 666.0
            final_copy['Tag 666 777'] = final_copy[col].map({666.0:1, 1.0:1, 
                                                             0:0, "*":0})
            final_copy = final_copy.drop(["Unnamed: 7"], axis=1)
            continue
        if col == 'Unnamed: 8':
            # Limpiamos caracteres no n√∫mericos, solo se encontr√≥ .
            # Limpiamos y reordenamos extra√±os inputs 666.0
            # Limpiamos y reordenamos extra√±os inputs 777.0
            final_copy['Tag 666 777'] = final_copy[col].map({666.0:1 , 777.0:2, 
                                                             0:0, ".":0, 1.0:1})
            final_copy = final_copy.drop(["Unnamed: 8"], axis=1)
            continue
        final_copy[col] = pd.to_numeric(final_copy[col])
final_copy.fillna(0,inplace = True)
for col in final_copy.columns:
    if col != 'Snippet': final_copy[col] = final_copy[col].astype(int)
print("Los √∫ltimos tipos son: " + str(final_copy.dtypes))
#Combinamos las columnas duplicadas
final_copy['Xenofobia'] = np.where(final_copy['Xenofobia'] == 0,
                                    np.where(final_copy['Autoxenofobia'] == 0, final_copy['autoxenofobia '], final_copy['Autoxenofobia']),
                                    final_copy['Xenofobia'])
final_copy['Positiva'] = np.where(final_copy['Positiva'] == 0,
                                    np.where(final_copy['Positivo'] == 0, final_copy['Positiva'], final_copy['Positivo']),
                                    final_copy['Positiva'])
final_copy = final_copy.drop(["autoxenofobia ", "Autoxenofobia", "Positivo"], axis=1)
```

```python
file1 = open("avance1.txt","r")
file2 = open("final1.txt","r")
ini = list()
fini = list()
for inicio in file1:
    inicio = inicio.strip()
    inicio = int(inicio)
    ini.append(inicio)
for inicio in file2:
    inicio = inicio.strip()
    inicio = int(inicio)
    fini.append(inicio)
nueva_base = pd.DataFrame()
for (posicion1, posicion2) in zip(ini, fini):
    nueva_base = pd.concat([nueva_base,segunda[posicion1 + 1 : posicion2]])
    print(posicion1, posicion2)
```
