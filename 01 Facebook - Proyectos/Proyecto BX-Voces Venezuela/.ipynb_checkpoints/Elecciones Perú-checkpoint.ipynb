```python
import numpy as np
import pandas as pd
import csv
df = pd.read_excel('Elecciones Perú .xlsx')
df
```

```python
df = df.fillna('')
df
```

```python
lista_paginas_facebook = []
lista_ciudades = df.columns
base = pd.DataFrame()
#.where(df[lista_ciudades[item]].str.contains('https', regex=True).notnull())
for item in range(len(lista_ciudades)):
    ciudad = df.where(df[lista_ciudades[item]].str.contains('https', regex=True).notnull())
    ciudad = ciudad[ciudad[lista_ciudades[item]] != '']
    ciudad = ciudad[ciudad[lista_ciudades[item]].str.contains("@")==False]
    ciudad[lista_ciudades[item]] = ciudad[lista_ciudades[item]].str.extract(r'\((.*?)\)', expand=False).fillna('')
    ciudad = ciudad[ciudad[lista_ciudades[item]].str.contains("http")==True]
    ciudad = ciudad[lista_ciudades[item]].fillna('')
    for index, value in ciudad.items():
        lista_paginas_facebook.append(value)
lista_paginas_facebook
```

```python
diccionario = {'Lista': lista_paginas_facebook}         
final = pd.DataFrame(diccionario)
final.to_csv('lista_elecciones.csv', index= False, header= False)
```
