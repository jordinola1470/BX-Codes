```python
# Listas:
df = df[(df['Medios - Medios alternativos'] == 'X') | (df['Medios - Medios tradicionales'] == 'X')]
df
```

```python
df['Page Type'].unique()
```

```python
medios = list()
medios_t = list()
medios_a = list()
tradicionales = open(r'D:\Trabajo\Bar√≥metro\BX\Base\medios.txt', 'r')
alternos = open(r'D:\Trabajo\Bar√≥metro\BX\Base\alternos.txt', 'r')
for i in tradicionales.readlines():
    medios.append(str(i).split('OR')[0].strip())
    medios_t.append(str(i).split('OR')[0].strip())
for i in alternos.readlines():
    medios.append(str(i).split('OR')[0].strip())
    medios_a.append(str(i).split('OR')[0].strip())
medios
```

```python
df = df[(df['Thread Author'].isin(medios)) | (df['Page Type'] == 'news') | (df['Page Type'] == 'blog') | (df['Page Type'] == 'forum') | (df['Page Type'] == 'tumblr') ]
df
```

```python
df = df[(df['Author'].isin(medios)) | (df['Page Type'] == 'news') | (df['Page Type'] == 'blog') | (df['Page Type'] == 'forum') | (df['Page Type'] == 'tumblr') ]
df
```

```python
df[(df['Medios - Medios alternativos'] == 'X') & (df['Medios - Medios tradicionales'] == 'X')].Domain.unique()
```
