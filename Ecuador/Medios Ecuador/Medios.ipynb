```python
import numpy as np
import pandas as pd
```

```python
medios = list()
medios1 = open('medios 41 - 80.txt','r')
medios1 = medios1.readlines()
for i in range(len(medios1)):
    medio = medios1[i].strip().split('@')[1]
    medios.append(medio)
medios1 = open('Medios 3.txt','r')
medios1 = medios1.readlines()
for i in range(len(medios1)):
    medio = medios1[i].strip()
    medios.append(medio)
medios1 = open('medios.txt','r')
medios1 = medios1.readlines()
for i in range(len(medios1)):
    medio = medios1[i].strip()
    medios.append(medio)
# function to get unique values
def unique(list1):    
    # initialize a null list
    unique_list = []
     
    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    return unique_list
final = unique(medios)
with open('final.txt', 'w') as f:
    for medio in final:
        f.write(medio + "\n") 
final
```
