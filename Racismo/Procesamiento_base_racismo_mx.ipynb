# Generaci√≥n de base para el mapa


```python
import numpy as np
import pandas as pd
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
```

```python
base_1 = pd.read_csv("mensajes_1_may_9_may.csv")
base_2 = pd.read_csv("mensajes_10_may_16_may.csv")
base_3 = pd.read_csv("mensajes_17_may_21_may.csv")
base_4 = pd.read_csv("mensajes_22_may_28_may.csv")
base_5 = pd.read_csv("mensajes_29_may_31_may.csv")
base_racismo_1 = base_1.append(base_2)
base_racismo_2 = base_3.append(base_4)
base_racismo_3 = base_racismo_1.append(base_racismo_2)
base_racismo_4 =  base_racismo_3.append(base_5)
base_racismo = base_racismo_4
base_racismo['Count'] = 1
base_racismo
```

```python
base_negativa_1 = pd.read_csv("mensajes_negativos_1_may_17_may.csv")
base_negativa_2 = pd.read_csv("mensajes_negativos_18_may_29_may.csv")
base_negativa_3 = pd.read_csv("mensajes_negativos_30_may_31_may.csv")
base_negativa_4 = base_negativa_1.append(base_negativa_2)
base_negativa_5 = base_negativa_4.append(base_negativa_3)
base_negativa = base_negativa_5
base_negativa['Count_negativo'] = 1
base_negativa.rename({"Followers": "Followers_negativo",	"Engagements": "Engagements_negativo", "Age":"Age_negativo", "Likes":"Likes_negativos",	
                      "Dislikes":"Dislikes_negativo",	"Comments":"Comments_negativo",	"Shares":"Shares_negativo"}, inplace=True, axis=1)
base_negativa
```

```python
base_agrupada = base_racismo.groupby('Location').sum()
base_agrupada['Splitted'] = base_agrupada.index.str.split(',')
base_agrupada.loc[:, 'Ciudad'] = base_agrupada.Splitted.map(lambda x: x[0])
try:
    base_agrupada.loc[:, 'Estado'] = base_agrupada.Splitted.str[1].str.strip()
    base_agrupada.loc[:, 'Pa√≠s'] = base_agrupada.Splitted.str[2].str.strip()
    base_agrupada.drop('Splitted', inplace=True, axis=1)
    base_agrupada_mexico_positiva = base_agrupada[base_agrupada['Pa√≠s'] == "Mexico"]
except:
    print("Do Nothing")
base_agrupada_mexico_positiva
```

```python
base_agrupada = base_negativa.groupby('Location').sum()
base_agrupada['Splitted'] = base_agrupada.index.str.split(',')
base_agrupada.loc[:, 'Ciudad'] = base_agrupada.Splitted.map(lambda x: x[0])
try:
    base_agrupada.loc[:, 'Estado'] = base_agrupada.Splitted.str[1].str.strip()
    base_agrupada.loc[:, 'Pa√≠s'] = base_agrupada.Splitted.str[2].str.strip()
    base_agrupada.drop('Splitted', inplace=True, axis=1)
    base_agrupada_mexico_negativa = base_agrupada[base_agrupada['Pa√≠s'] == "Mexico"]
except:
    print("Do Nothing")
base_agrupada_mexico_negativa
```

```python
base_agrupada_estados_mexico_positiva = base_agrupada_mexico_positiva.groupby("Estado").sum()
base_agrupada_estados_mexico_negativa = base_agrupada_mexico_negativa.groupby("Estado").sum()
base_agrupada_estados_mexico = pd.merge(left=base_agrupada_estados_mexico_negativa, right=base_agrupada_estados_mexico_positiva, right_on=base_agrupada_estados_mexico_positiva.index , left_on=base_agrupada_estados_mexico_negativa.index, how='inner')
base_agrupada_estados_mexico.rename({'key_0':'Estado'}, inplace=True, axis=1)
base_agrupada_estados_mexico
```

```python
from sklearn import preprocessing
normalized_arr = preprocessing.normalize(base_agrupada_estados_mexico['Engagements'].to_numpy().reshape(1,-1))
rounded_arr = lambda x: round(x,4)
vfunc = np.vectorize(rounded_arr)
vfunc(normalized_arr)
```

```python
scaler = preprocessing.MinMaxScaler()
base_agrupada_estados_mexicos_counts = base_agrupada_estados_mexico[['Estado','Count', 'Count_negativo']]
base_agrupada_estados_mexicos_counts.rename({'key_0': 'Estado'}, axis=1, inplace=True)
base_agrupada_estados_mexicos_counts.set_index('Estado', inplace=True)
base_agrupada_estados_mexicos_counts['% Odio'] = base_agrupada_estados_mexicos_counts['Count_negativo']/base_agrupada_estados_mexicos_counts['Count']
base_agrupada_estados_mexicos_counts.to_csv("Base_estados_mexicanos_completa.csv")
base_agrupada_estados_mexicos_counts
```

```python
names = base_agrupada_estados_mexicos_counts.columns
d = scaler.fit_transform(base_agrupada_estados_mexicos_counts)
scaled_df = pd.DataFrame(d, columns=names)
scaled_df['Estado'] = base_agrupada_estados_mexico.Estado
scaled_df.set_index('Estado', inplace=True)
scaled_df.head(8)
```

```python
fulled_df = pd.merge(left= scaled_df, right= base_agrupada_estados_mexico, left_on=scaled_df.index, right_on=base_agrupada_estados_mexico.Estado, how="inner")
fulled_df.rename({'key_0':'Estado', 'Followers_x':'Followers_n'
                 , 'Engagements_x':'Engagements_n', 'Count_x':'Count_n'
                 , 'Count_y':'Count', 'Followers_y':'Followers', 'Engagements_y':'Engagements', 'Count_negativo_y':'Count_negativo'}, inplace=True, axis=1)
fulled_df.columns
```

```python
fulled_df[['Count_negativo', 'Count_negativo_x', '% Odio', 'Count']]
```

```python
scaled_df.to_excel("Base_estados_mexicanos_normalizada.xlsx")
scaled_df.to_csv("Base_estados_mexicanos_normalizada.csv")
```
