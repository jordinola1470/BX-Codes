```python
import urllib.request, urllib.parse, urllib.error
from urllib.request import urlopen
from bs4 import BeautifulSoup
import ssl
#soup1 = BeautifulSoup("<p>Some<b>bad<i>HTML")
#print(soup1.prettify())
```

```python
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
url = input('Enter - ')
html = urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, "html.parser")
#Retrive an span tag <span data structure ...> value </span>
ans = 0
tags = soup('span')
for tag in tags:
    print('TAG:',tag)
    print('URL:',tag.get('href', None)) 
    print('Contents:',tag.contents[0]) 
    print('Attrs:',tag.attrs)
    ans += int(tag.contents[0])
print('Answer',ans)
```

```python
url = input('Enter URL - ')
count = int(input('Enter count: '))
position = int(input('Enter position: '))
html = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, 'html.parser')

# Retrieve all of the anchor tags <a href=...>
centinel1 = 0
centinel = 0
tags = soup('a')
print('Retrieving: ',url)
for tag in tags:
    if centinel1 == count:
        break
    if centinel == position-1:
        url = tag.get('href',None)
        print('Retrieving: ',url)
        html = urllib.request.urlopen(url, context=ctx).read()
        soup = BeautifulSoup(html, 'html.parser')
        tags = soup('a')
        centinel = 0
        centinel1 += 1
        for tag in tags:
            if centinel1 == count:
                break
            if centinel == position-1:
                url = tag.get('href',None)
                print('Retrieving: ',url)
                html = urllib.request.urlopen(url, context=ctx).read()
                soup = BeautifulSoup(html, 'html.parser')
                tags = soup('a')
                centinel = 0
                centinel1 += 1
                for tag in tags:
                    if centinel1 == count:
                        break
                    if centinel == position-1:
                        url = tag.get('href',None)
                        print('Retrieving: ',url)
                        html = urllib.request.urlopen(url, context=ctx).read()
                        soup = BeautifulSoup(html, 'html.parser')
                        tags = soup('a')
                        centinel = 0
                        centinel1 += 1
                        for tag in tags:
                            if centinel1 == count:
                                break
                            if centinel == position-1:
                                url = tag.get('href',None)
                                print('Retrieving: ',url)
                                html = urllib.request.urlopen(url, context=ctx).read()
                                soup = BeautifulSoup(html, 'html.parser')
                                tags = soup('a')
                                centinel = 0
                                centinel1 += 1
                            centinel += 1
                    centinel += 1
            centinel += 1
    centinel += 1
```

```python
# Recursiòn para varios casos
url = input('Enter URL - ')
count = int(input('Enter count: '))
position = int(input('Enter position: '))
centinel = 0
i = 0
html = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, 'html.parser')
tags = soup('a')
print('Retrieving: ',url)
while i < count:
    for tag in tags:
        if centinel == position-1:
            url = tag.get('href',None)
            print('Retrieving: ',url)
            html = urllib.request.urlopen(url, context=ctx).read()
            soup = BeautifulSoup(html, 'html.parser')
            tags = soup('a')
            centinel = 0
            i += 1
            break
        centinel += 1
```
